```python
# Adım 1: Gerekli kütüphaneler

import pandas as pd
from sentence_transformers import SentenceTransformer, util
import torch

# Adım 2: CSV dosyasını oku
df = pd.read_csv("000dialog.csv")  # Dosya adı doğruysa çalışır

# Adım 3: Modeli yükle (hızlı ve iyi sonuç verir)
model = SentenceTransformer('paraphrase-MiniLM-L6-v2')

# Adım 4: Soru ve cevapları listeliyoruz
sorular = df["Soru"].tolist()
cevaplar = df["Cevap"].tolist()

# Adım 5: Tüm soruların vektör temsillerini al
soru_embeddings = model.encode(sorular, convert_to_tensor=True)

# Adım 6: Kullanıcı sorusuna en yakın cevabı bul
def cevap_ver(kullanici_sorusu):
    soru_embedding = model.encode(kullanici_sorusu, convert_to_tensor=True)
    benzerlik = util.cos_sim(soru_embedding, soru_embeddings)
    en_yakin = torch.argmax(benzerlik)
    return cevaplar[en_yakin]

# Adım 7: Test sistemi (soru sor, cevap al)
while True:
    giris = input("👤 Soru yaz (çıkmak için q): ")
    if giris.lower() == 'q':
        break
    print("🤖 Cevap:", cevap_ver(giris))
```

```python
!pip install --upgrade torchvision torchaudio --index-url https://download.pytorch.org/whl/cu118


# Kütüphane içe aktarma
import pandas as pd  # Veri işlemleri için
import torch  # PyTorch kullanımı için
import fitz  # PyMuPDF ile PDF okuma
import gradio as gr  # Web arayüzü oluşturmak için
from sentence_transformers import SentenceTransformer, util  # Cümle embedding için
from gtts import gTTS  # Google Text-to-Speech (metni sese çevirir)
from google.colab import files  # Colab arayüzünden dosya yüklemek için

# 📁 1. CSV dosyası yükleniyor
print("📌 Önce CSV dosyasını yükleyin (örn: dialog.csv)")
csv_uploaded = files.upload()
csv_path = list(csv_uploaded.keys())[0]  # Yüklenen CSV dosyasının adını al

#  2. PDF dosyası yükleniyor
print("📌 Şimdi PDF dosyasını yükleyin (örn: ornderin.pdf)")
pdf_uploaded = files.upload()
pdf_path = list(pdf_uploaded.keys())[0]  # Yüklenen PDF dosyasının adını al

#  Transformer model yükleniyor (MiniLM tabanlı)
model = SentenceTransformer('paraphrase-MiniLM-L6-v2')

#  CSV verisi okunuyor
df = pd.read_csv(csv_path)
sorular = df["Soru"].tolist()     # CSV'deki sorular listeleniyor
cevaplar = df["Cevap"].tolist()   # CSV'deki cevaplar listeleniyor
soru_embeddings = model.encode(sorular, convert_to_tensor=True)  # Sorular embedding'e çevriliyor

#  PDF'ten metin çıkarılıyor
def pdf_metnini_ayikla(pdf_path):
    doc = fitz.open(pdf_path)  # PDF dosyasını aç
    tam_metin = ""
    for sayfa in doc:
        tam_metin += sayfa.get_text("text") + "\n"  # Her sayfayı metin olarak ekle
    # Paragrafları boş satırlarla ayır, 30 karakterden kısa olanları alma
    paragraflar = [p.strip() for p in tam_metin.split("\n\n") if len(p.strip()) > 30]
    return paragraflar

# PDF paragrafları çıkarılıyor ve embedding'e çevriliyor
pdf_paragraflar = pdf_metnini_ayikla(pdf_path)
pdf_embeddings = model.encode(pdf_paragraflar, convert_to_tensor=True)
print(f"📄 {len(pdf_paragraflar)} paragraf yüklendi.")

#  Geçmiş konuşmaları tutmak için liste
gecmis = []

# Soruya karşılık en iyi cevabı bulan fonksiyon
def cevap_ver(kullanici_sorusu):
    soru_embedding = model.encode(kullanici_sorusu, convert_to_tensor=True)

    # CSV verisiyle benzerlik karşılaştırması
    benzerlik_csv = util.cos_sim(soru_embedding, soru_embeddings)
    en_yakin_csv = torch.argmax(benzerlik_csv)
    skor_csv = benzerlik_csv[0][en_yakin_csv].item()

    # PDF verisiyle benzerlik karşılaştırması
    benzerlik_pdf = util.cos_sim(soru_embedding, pdf_embeddings)
    en_yakin_pdf = torch.argmax(benzerlik_pdf)
    skor_pdf = benzerlik_pdf[0][en_yakin_pdf].item()

    # Cevabı seç (yüksek skorlu olan)
    if skor_csv >= skor_pdf and skor_csv > 0.5:
        return cevaplar[en_yakin_csv]
    elif skor_pdf > 0.5:
        cevap = pdf_paragraflar[en_yakin_pdf].replace("\n", " ").strip()
        return cevap[:1000] + "..." if len(cevap) > 1000 else cevap
    else:
        return "Üzgünüm, bu konuyla ilgili net bir cevabım yok."

# Gradio üzerinden çağrılan chatbot fonksiyonu
def chatbot_gradio(soru):
    yanit = cevap_ver(soru)
    gecmis.append(f"👤 {soru}\n🤖 {yanit}")

    # Metni ses dosyasına çevir
    tts = gTTS(yanit, lang='tr')
    tts.save("cevap.mp3")

    # Son 5 konuşma ve sesli yanıt döndürülüyor
    return "\n\n".join(gecmis[-5:]), "cevap.mp3"

#  Arayüz teması
custom_theme = gr.themes.Monochrome(
    primary_hue="slate",
    neutral_hue="stone",
    font=["Fira Code", "monospace"]
)
"""
#  Gradio arayüzü kurulumu
demo = gr.Interface(
    fn=chatbot_gradio,
    inputs=gr.Textbox(lines=2, placeholder="Sorunuzu yazın...", label="Soru"),
    outputs=[
        gr.Textbox(lines=10, label="Sohbet Geçmişi"),   # Sohbet geçmişi kutusu
        gr.Audio(label="Sesli Cevap", autoplay=False)   # Sesli cevap kutusu
    ],
    title="SMARTGPTBOT 🤖",
    description=" Aklınızdaki sorulara cevap veriyorum. Cevapları sesli de okuyabilirim.",
    theme=custom_theme
)
"""



demo = gr.Interface(
    fn=chatbot_gradio,
    inputs=gr.Textbox(lines=2, placeholder="Sorunuzu yazın...", label="Soru"),
    outputs=[
        gr.Textbox(lines=10, label="Sohbet Geçmişi"),
        gr.Audio(label="Sesli Cevap", autoplay=False)
    ],
    title="SMARTGPTBOT 🤖",
    description="Aklınızdaki sorulara cevap veriyorum. Cevapları sesli de okuyabilirim.",
    theme=custom_theme,
    submit_btn="Gönder",   # Submit yerine "Gönder"
    clear_btn="Temizle"    # Clear yerine "Temizle"
)

#  Arayüz başlatılıyor
demo.launch()
```
